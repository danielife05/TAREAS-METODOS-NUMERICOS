5. La funci√≥n $f(x) = \tan{\pi x} - 6$ tiene cero en $(\frac{1}{\pi})$ arcotangente $6 \approx 0.447431543$. Sea $p_0 = 0$ y $p_1 = 0.48$ y use 10 iteraciones en cada uno de los siguientes m√©todos para aproximar esta ra√≠z. ¬øCu√°l m√©todo es m√°s eficaz y por qu√©?

a. M√©todo de Bisecci√≥n

```python
def sign(x: float) -> int:
    if x > 0:
        return 1
    elif x < 0:
        return -1
    else:
        return 0
    
from typing import Callable

def bisection(
    a: float, b: float, *, equation: Callable[[float], float], N: int
) -> tuple[float, int] | None:
    i = 1
    
    assert a < b, "a not lower than b, the interval is not valid."

    assert (
        equation(a) * equation(b) < 0
    ), "The function does not change sign over the interval."

    Fa = equation(a)
    p = a 
    for i in range(N + 1):
        p = a + (b - a) / 2
        FP = equation(p)
        
        if sign(Fa) * sign(FP) > 0:
            a = p
            Fa = FP
            
        else:
            b = p

    return p, i

p, i = bisection(a = 0, b = 0.48, 
                 equation =  lambda x : np.tan(np.pi*x) - 6,
                N = 10)
print(f"Ra√≠z resultante: {p} en {i} iteraciones")
```

b. M√©todo de Newton

Derivada de la funci√≥n:

$\pi \sec^2{\pi x}$

```python
try:
    p1 = newton(func=lambda x: np.tan(np.pi * x) - 6, x0=0, 
                fprime=lambda x: np.pi * (1 / np.cos(np.pi * x))**2, maxiter=10, x1=0.48)
    print(p1)
except RuntimeError:
    print("No se logr√≥ la convergencia despu√©s de 10 iteraciones.")
```

c. M√©todo de la Secante

```python
try:
    p1 = newton(func=lambda x: np.tan(np.pi * x) - 6, x0=0, maxiter=10, x1=0.48)
    print(p1)
except RuntimeError as e:
    print("No se logr√≥ la convergencia despu√©s de 10 iteraciones.")
```

El m√©todo de bisecci√≥n es m√°s efectivo para encontrar una ra√≠z en la funci√≥n. Los otros dos m√©todos dependen de la pendiente o de una aproximaci√≥n que corta el eje $x$, lo cual puede llevar a resultados fuera del intervalo especificado. Adem√°s, la funci√≥n no es continua en todos los n√∫meros reales, lo que contribuye a las dificultades de convergencia con estos m√©todos.
