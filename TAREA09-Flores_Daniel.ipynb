a) $x_1 + 2x_2 = 0$, $x_1 - x_2 = 0$

```python
import numpy as np
import matplotlib.pyplot as plt

points_x = [0]
points_y = [0]

x = np.linspace(-10, 10, 500)
y1 = - x / 2
y2 = x      

plt.figure(figsize=(8, 6))
plt.plot(x, y1, label=r"$x_1 + 2x_2 = 0$", color="orange")
plt.plot(x, y2, label=r"$x_1 - x_2 = 0$", color="red")
plt.scatter(points_x, points_y, color="k", label="Intersección", zorder=5)
plt.axhline(0, color='black', linewidth=0.8)
plt.axvline(0, color='black', linewidth=0.8)
plt.title("Sistema de Ecuaciones Lineales", fontsize=14)
plt.grid(alpha=0.4)
plt.legend(fontsize=12)
plt.show()
```

Las rectas no son paralelas ni coincidentes totalmente, por lo que tienen exactamente un punto de intersección, por lo que se cruzan en el punto $(0,0)$.

Este sistema de ecuaciones no tiene una solución porque las funciones son paralelas, por lo que nunca tendrán un punto de intersección.

c) $2x_1 + x_2 = -1$, $x_1 + x_2 = 2$, $x_1 - 3x_2 = 5$

```python
y1 = - 2 * x - 1
y2 = 2 - x
y3 = (x - 5) / 3     

points_x = [-3, 2/7, 11/4]
points_y = [5, -11/7, -3/4] 

plt.figure(figsize=(8, 6))
plt.plot(x, y1, label=r"$2x_1 + x_2 = -1$", color="orange")
plt.plot(x, y2, label=r"$x_1 + x_2 = 2$", color="red")
plt.plot(x, y3, label=r"$x_1 - 3x_2 = 5$", color="blue")
plt.scatter(points_x, points_y, color="k", label="Intersección", zorder=5)
plt.axhline(0, color='black', linewidth=0.8)
plt.axvline(0, color='black', linewidth=0.8)
plt.title("Sistema de Ecuaciones Lineales", fontsize=14)
plt.grid(alpha=0.4)
plt.legend(fontsize=12)
plt.show()
```

d) $2x_1 + x_2 + x_3 = 1$, $2x_1 + 4x_2 - x_3 = -1$

```python
def plot_system_d():
    from mpl_toolkits.mplot3d import Axes3D

    fig = plt.figure()
    ax = fig.add_subplot(111, projection='3d')

    x = np.linspace(-10, 10, 10)
    y = np.linspace(-10, 10, 10)
    X, Y = np.meshgrid(x, y)

    Z1 = 1 - 2*X - Y
    Z2 = -1 - 2*X - 4*Y

    ax.plot_surface(X, Y, Z1, alpha=0.5, cmap='viridis', edgecolor='none', label="$2x_1 + x_2 + x_3 = 1$")
    ax.plot_surface(X, Y, Z2, alpha=0.5, cmap='plasma', edgecolor='none', label="$2x_1 + 4x_2 - x_3 = -1$")

    ax.set_title("Sistema d)")
    ax.set_xlabel("$x_1$")
    ax.set_ylabel("$x_2$")
    ax.set_zlabel("$x_3$")
    plt.show()

plot_system_d()
```

### a)

,$\frac{5}{3}x_1+\frac{2}{3}x_2+\frac{2}{3}x3=1$

,$2x_1+x_2+4x_3=11$

```python
import numpy as np

Ab = np.array([
    [-1, 4, 1, 8],
    [5/3, 2/3, 2/3, 1],
    [2, 1, 4, 11]
], dtype=float)

A = Ab[:, :-1] 
b = Ab[:, -1] 
solucion = np.linalg.solve(A, b)

print("La solución del sistema es:")
print(f"x1 = {solucion[0]:.2f}, x2 = {solucion[1]:.2f}, x3 = {solucion[2]:.2f}")
```

```python
Ab = np.array([
    [4,2,-1,-5],
    [1/9,1/9,1/3,-1],
    [1,4,2,9]
], dtype=float)

A = Ab[:, :-1] 
b = Ab[:, -1] 
solucion = np.linalg.solve(A, b)

print("La solución del sistema es:")
print(f"x1 = {solucion[0]:.2f}, x2 = {solucion[1]:.2f}, x3 = {solucion[2]:.2f}")
```

### Problema 3: Resolución de sistemas lineales mediante eliminación gaussiana

Utilice el algoritmo de eliminación gaussiana para resolver, de ser posible, los siguientes sistemas lineales, y determine si se necesitan intercambios de fila:


a)
$x_1 - x_2 + 3x_3 = 2$

$3x_1 - 3x_2 + x_3 = -1$

$x_1 + x_2 = 3$

```python
import numpy as np

def resolver_sistema(Ab):
    A = Ab[:, :-1]
    b = Ab[:, -1]
    try:
        solucion = np.linalg.solve(A, b)
        return solucion
    except np.linalg.LinAlgError as e:
        return f"Error: {e}"

Ab_a = np.array([
    [1, -1, 3, 2],
    [3, -3, 1, -1],
    [1, 1, 0, 3]
], dtype=float)

solucion_a = resolver_sistema(Ab_a)
print("Solución sistema:", solucion_a)
```

b) $2x_1 - 1.5x_2 + 3x_3 = 1$

$-x_1 + 2x_3 = 3$

$4x_1 - 4.5x_2 + 5x_3 = 1$

```python
Ab_b = np.array([
    [2, -1.5, 3, 1],
    [-1, 0, 2, 3],
    [4, -4.5, 5, 1]
], dtype=float)

solucion_b = resolver_sistema(Ab_b)
print("Solución sistema:", solucion_b)
```

c) $2x_1 = 3$

$x_1 + 1.5x_2 = 4.5$

$-3x_2 + 0.5x_3 = -6.6$

$2x_1 - 2x_2 + x_4 = 0.8$

```python
Ab_c = np.array([
    [2, 0, 0, 0, 3],
    [1, 1.5, 0, 0, 4.5],
    [0, -3, 0.5, 0, -6.6],
    [2, -2, 0, 1, 0.8]
], dtype=float)

solucion_c = resolver_sistema(Ab_c)
print("Solución sistema:", solucion_c)
```

d) $x_1 + x_2 + x_4 = 2$

$2x_1 - x_2 - x_3 + x_4 = 1$

$4x_1 - x_2 - 2x_3 - x_4 = 0$

$3x_1 - x_2 - x_3 + 2x_4 = -3$

```python
Ab_d = np.array([
    [1, 1, 0, 1, 2],
    [2, -1, -1, 1, 1],
    [4, -1, -2, -1, 0],
    [3, -1, -1, 2, -3]
], dtype=float)

solucion_d = resolver_sistema(Ab_d)
print("Solución sistema:", solucion_d)
```

Para garantizar que el sistema de ecuaciones nno tenga soluciones primero realizaremos una reduccón del sistema de ecuaciones a través del método de Gauss.


$$ \begin{bmatrix}
1 & -1 & \alpha & -2\\
-1 & 2 & -\alpha & 3\\
\alpha & 1 & 1 & 2
 \end{bmatrix} \equiv F1+F2->F2\begin{bmatrix}
 1 & -1 & -\alpha & -2\\
 0 & 1 & 0 & 1\\
 \alpha & 1 &1 & 2
 \end{bmatrix}\\
 \equiv (-x-1)F2+F3 ->F3 \begin{bmatrix}
 1 & -1 & -\alpha & -2\\
  0 & 1 & 0 & 1\\
  0 & 0 & -\alpha^2+1 & \alpha+1
 \end{bmatrix}$$ 

de este modo para que el sistema no tenga solución $-\alpha^2+1 = 0$ mientras que $\alpha+1\neq 0$, a continuación haremos cumplir dichas condiciones:
$$-\alpha^2+1=0\\-\alpha^2=-1\\
\alpha^2=1\\
\alpha=\pm1$$

\alpha\neq-1$$
por lo tanto $$\alpha=1$$
